```python
import os
import shutil
import random

sample_size = 5

random_seed = 36
random.seed(random_seed)

src_root = '../datasets/DTD235/images'
dst_root = '../datasets/DTD235/original'

if not os.path.exists(dst_root):
    os.makedirs(dst_root)

folders = [
    'chequered', 'cobwebbed', 'cracked', 'crosshatched', 'crystalline', 'dotted', 'fibrous', 'flecked',
    'freckled', 'frilly', 'gauzy', 'grid', 'grooved', 'honeycombed', 'interlaced', 'knitted', 'lacelike',
    'lined', 'marbled', 'matted', 'meshed', 'paisley', 'perforated', 'pitted', 'pleated', 'polka-dotted',
    'porous', 'potholed', 'scaly', 'smeared', 'spiralled', 'sprinkled', 'stained', 'stratified', 'striped',
    'studded', 'swirly', 'veined', 'waffled', 'woven', 'wrinkled', 'zigzagged', 'banded', 'blotchy', 'braided',
    'bubbly', 'bumpy'
]

for folder in folders:
    src_folder = os.path.join(src_root, folder)
    image_files = os.listdir(src_folder)
    sampled_images = random.sample(image_files, min(sample_size, len(image_files)))
    print(f'sampling {sampled_images} images from {folder}...')
    for image in sampled_images:
        src_file = os.path.join(src_folder, image)
        dst_file = os.path.join(dst_root, image)
        shutil.copy2(src_file, dst_file)

print("dtd dataset is ready for testing.")
```
